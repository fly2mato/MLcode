```python
from sklearn.svm import LinearSVC
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
    
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
X = X[y<2, :2]
y = y[y<2]

plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
    
    
```

```python
def plot_decision_boundary(model, axis):    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1,1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1,1)
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]
    
    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)
    
    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, cmap=custom_cmap)

from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

standardScaler = StandardScaler()
standardScaler.fit(X)
X_standard = standardScaler.transform(X)


svc = LinearSVC(C=10**9)
svc.fit(X_standard, y*2-1)

plot_decision_boundary(svc, axis=[-3, 3, -3, 3])
plt.scatter(X_standard[y==0, 0], X_standard[y==0, 1], color='red')
plt.scatter(X_standard[y==1, 0], X_standard[y==1, 1], color='blue')

svc.coef_, svc.intercept_

xx0 = np.linspace(-3,3,100)
xx1 = (1-svc.intercept_-svc.coef_[0,0]*xx0)/svc.coef_[0,1]
xx1 = np.clip(xx1, -3,3)
plt.plot(xx0, xx1, color='red')
xx1 = (-1-svc.intercept_-svc.coef_[0,0]*xx0)/svc.coef_[0,1]
xx1 = np.clip(xx1, -3,3)
plt.plot(xx0, xx1, color='red')

plt.show()

svc.coef_, svc.intercept_
```

```python
class LinearKernel(object):
    def __call__(self, x, y):
        return np.dot(x, y.T)

class PolyKernel(object):
    def __init__(self, p=2):
        self.p = p
    def __call__(self, x, y):
        return (x.dot(y.T)+1)**self.p

class SigmoidKernel(object):
    def __init__(self, gamma=1):
        self.gamma=gamma
        
    def __call__(self, x, y):
        return np.tanh(self.gamma*(x.dot(y.T))+1)

#RBF something wrong
'''    
class RBF(object):
    def __init__(self, gamma=1):
        self.gamma=gamma
    def __call__(self, x, y):
        return -self.gamma * np.exp(np.linalg.norm(x-y)**2)
'''
    
class svm(object):
    def __init__(self, X, y, C, tol=1e-6, kernel=LinearKernel()):
        self.X = X
        self.y = y
        self.C = C
        self.alpha = np.zeros(y.shape)
        self.index = np.arange(y.size)
        self.tol = tol
        self.eps = 1e-3
        self.kernel = kernel

        self.b = 0 
        
        self.K = np.zeros([self.X.shape[0], self.X.shape[0]])
        for i in range(X.shape[0]):
            for j in range(i, X.shape[0]):
                self.K[i,j] = self.K[j,i] = self.kernel(X[i,:],X[j,:])
        
        self.E = np.zeros(y.shape)
        for i in range(X.shape[0]):
            self.E[i] = self.calcE(i)
        

    
    def calcE(self, i):
        E = (self.alpha*self.y).dot(self.K[i,:])+ self.b - self.y[i]
        return E
        
    def takeStep(self,i1,i2):
        alpha1, alpha2 = self.alpha[[i1,i2]]
        y1, y2 = self.y[[i1,i2]]
        E1, E2 = self.E[[i1,i2]]
        s = y1*y2
        if (y1 != y2):
            L = max(0, alpha2 - alpha1)
            H = min(self.C, self.C + alpha2 - alpha1)
        else :
            L = max(0, alpha2 + alpha1 - self.C)
            H = min(self.C, alpha2 + alpha1) 
        if L==H:
            return 0
        
        K11 = self.K[i1,i1]
        K12 = self.K[i1,i2]
        K22 = self.K[i2,i2]
        eta = K11+K22-2*K12
        
        if eta>0:
            a2 = np.clip(alpha2 + y2*(E1-E2)/eta,a_min=L,a_max=H)
        else:
            f1 = y1*(E1+self.b)-alpha1*K11-s*alpha2*K12
            f2 = y2*(E2+self.b)-s*alpha1*K12-alpha2*K22
            L1 = alpha1 + s*(alpha2-L)
            H1 = alpha1 + s*(alpha2-H)
            Lobj = L1*f1+L*f2+.5*L1**2*K11+.5*L**2*K22+s*L*L1*K12
            Hobj = H1*f1+H*f2+.5*H1**2*K11+.5*H**2*K22+s*H*H1*K12
            if Lobj < Hobj-self.eps:
                a2=L
            elif Lobj > Hobj+self.eps:
                a2=H
            else:
                a2 = alpha2
        
        
        if abs(a2-alpha2) < self.eps*(a2+alpha2+self.eps):
            return 0
        
        a1 = alpha1+s*(alpha2-a2)
        
        #Platt论文里面用的y=wx-b，所以这里b的更新时需要反号
        b1 = -(E1+y1*(a1-alpha1)*K11+y2*(a2-alpha2)*K12)+self.b
        b2 = -(E2+y1*(a1-alpha1)*K12+y2*(a2-alpha2)*K22)+self.b
        if a1>0 and a1<self.C:
            self.b = b1
        elif a2>0 and a2<self.C:
            self.b = b2
        else:
            self.b = (b1+b2)/2
    
        self.alpha[i1] = a1
        self.alpha[i2] = a2
            
        for i in range(X.shape[0]):
            self.E[i] = self.calcE(i)

        return 1
    
    def examineExample(self, i2):
        y2 = self.y[i2]
        alpha2 = self.alpha[i2]
        E2 = self.E[i2]
        r2 = E2*y2
        if (r2<-self.tol and alpha2<self.C) or (r2>self.tol and alpha2>0) :
            
            if ((self.alpha>0) & (self.alpha<self.C)).sum() > 1:
                if E2 > 0:
                    i1 = self.E.argmin()
                else :
                    i1 = self.E.argmax()

                if (i1 != i2) and self.takeStep(i1,i2):
                    return 1 
                
            cindex = self.index[(self.alpha>0) & (self.alpha<self.C)]
            if len(cindex)>0:                
                np.random.shuffle(cindex)
                for i1 in cindex:
                    if (i1 != i2) and self.takeStep(i1,i2):
                        return 1
                
            cindex = self.index.copy()          
            np.random.shuffle(cindex)
            for i1 in cindex:
                if (i1 != i2) and self.takeStep(i1,i2):
                    return 1
        return 0
        
    def fit(self):
        numChanged = 0
        examineAll = 1
        while numChanged>0 or examineAll:
            numChanged=0
            if examineAll:
                for i in self.index:
                    numChanged += self.examineExample(i)
            else:
                cindex = self.index[(self.alpha>0) & (self.alpha<self.C)]
                if len(cindex)>0:
                    for i in cindex:
                        numChanged += self.examineExample(i)

            if examineAll==1:
                examineAll=0
            elif numChanged==0:
                examineAll=1
                
        ya = (self.y*self.alpha).reshape(1,-1)
        w = ya.dot(self.X).reshape(-1)
        self.w = w
        return w,self.b
    
    def predict(self, X):
        '''
        K = np.zeros(X.shape[0])
        for j in range(X.shape[0]):
            for i in range(self.X.shape[0]):
                K[j] = self.kernel(self.X[i], X[j])        
        '''
        y = (self.alpha*self.y).dot(self.kernel(self.X, X)) + self.b
        #y = (self.alpha*self.y).dot(K) + self.b
        y[y>0] = 1
        y[y<0] = -1
        return y
        
```
