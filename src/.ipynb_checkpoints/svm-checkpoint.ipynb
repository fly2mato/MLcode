```python
from sklearn.svm import LinearSVC
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

X, y = make_classification(n_samples=1000,n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)
clf = LinearSVC(random_state=0, tol=1e-5)
clf.fit(X, y)

print(clf.coef_)
print(clf.intercept_)
print(clf.predict([[0, 0]]))

plt.figure();
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y)
plt.show()

y=y*2-1
```

```python
class svm(object):
    def __init__(self, X, y, C, tol=1e-3):
        self.X = X
        self.y = y
        self.C = C
        self.alpha = np.zeros(y.shape)
        #self.alpha = np.random.random(y.shape)
        self.index = np.arange(y.size)
        self.tol = tol
        self.eps = 1e-3
        
        ya = (self.y*self.alpha).reshape(1,-1)
        self.w = ya.dot(X).reshape(-1)
        
        self.b = self.w.dot(X[0,:]) - y[0]
        self.E = -y+self.b
        
    def kernel(self, i1, i2):
        K = X[i1,:].dot(X[i2,:])
        return K
    
    def calcE(self, i):
        E = 0
        for j in self.index[(self.alpha>0) & (self.alpha<self.C)]:
            E += y[j]*self.alpha[j]*self.kernel(i,j)
        E += self.b-y[i] 
        return E
        
    def takeStep(self,i1,i2):
        alpha1, alpha2 = self.alpha[[i1,i2]]
        y1, y2 = self.y[[i1,i2]]
        E1, E2 = self.E[[i1,i2]]
        s = y1*y2
        if (y1 != y2):
            L = max(0, alpha2 - alpha1)
            H = min(self.C, self.C + alpha2 - alpha1)
        else :
            L = max(0, alpha2 + alpha1 - self.C)
            H = min(self.C, alpha2 + alpha1) 
        if L==H:
            return 0
        
        K11 = self.kernel(i1,i1)
        K12 = self.kernel(i1,i2)
        K22 = self.kernel(i2,i2)
        eta = K11+K22-2*K12
        
        a2 = np.clip(alpha2 + y2*(E1-E2)/eta,a_min=L,a_max=H)
        '''
        if eta>0:
            a2 = np.clip(alpha2 + y2*(E1-E2)/eta,a_min=L,a_max=H)
        else:
            f1 = y1*(E1+self.b)-alpha1*K11-s*alpha2*K12
            f2 = y2*(E2+self.b)-s*alpha1*K12-alpha2*K22
            L1 = alpha1 + s*(alpha2-L)
            H1 = alpha1 + s*(alpha2-H)
            Lobj = L1*f1+L*f2+.5*L1**2*K11+.5*L**2*K22+s*L*L1*K12
            Hobj = H1*f1+H*f2+.5*H1**2*K11+.5*H**2*K22+s*H*H1*K12
            if Lobj < Hobj-self.eps:
                a2=L
            elif Lobj > Hobj+self.eps:
                a2=H
            else:
                a2 = alpha2
        '''
        
        if abs(a2-alpha2) < self.eps*(a2+alpha2+self.eps):
            return 0
        
        a1 = alpha1+s*(alpha2-a2)
        
        #if a1>0 and a1<self.C:
        b1 = -(E1+y1*(a1-alpha1)*K11+y2*(a2-alpha2)*K12)+self.b
        #if a2>0 and a2<self.C:
        b2 = -(E2+y1*(a1-alpha1)*K12+y2*(a2-alpha2)*K22)+self.b
        self.b = (b1+b2)/2
        
        self.w += y1*(a1-alpha1)*X[i1,:] + y2*(a2-alpha2)*X[i2,:]
            
        self.E[i1] = self.calcE(i1)
        self.E[i2] = self.calcE(i2)
        self.alpha[i1] = a1
        self.alpha[i2] = a2
        return 1
    
    def examineExample(self, i2):
        y2 = self.y[i2]
        alpha2 = self.alpha[i2]
        E2 = self.E[i2]
        r2 = E2*y2
        if (r2<-self.tol and alpha2<self.C) or (r2>self.tol and alpha2>0) :
            if ((self.alpha>0) & (self.alpha<self.C)).sum() > 1:
                
                if E2 > 0:
                    i1 = self.E.argmin()
                else :
                    i1 = self.E.argmax()

                if (i1 != i2) and self.takeStep(i1,i2):
                    return 1 
                
            cindex = self.index[(self.alpha>0) & (self.alpha<self.C)]
            if len(cindex)>0:                
                np.random.shuffle(cindex)
                for i1 in cindex:
                    if (i1 != i2) and self.takeStep(i1,i2):
                        return 1
                
            cindex = self.index.copy()          
            np.random.shuffle(cindex)
            for i1 in cindex:
                if (i1 != i2) and self.takeStep(i1,i2):
                    return 1
        return 0
        
    
    def fit(self):
        numChanged = 0
        examineAll = 1
        while numChanged>0 or examineAll:
            numChanged=0
            if examineAll:
                for i in self.index:
                    numChanged += self.examineExample(i)
            else:
                cindex = self.index[(self.alpha>0) & (self.alpha<self.C)]
                if len(cindex)>0:
                    for i in cindex:
                        numChanged += self.examineExample(i)

            if examineAll==1:
                examineAll=0
            elif numChanged==0:
                examineAll=1
                
        ya = (self.y*self.alpha).reshape(1,-1)
        w = ya.dot(X).reshape(-1)
        return w,self.b
        
```
